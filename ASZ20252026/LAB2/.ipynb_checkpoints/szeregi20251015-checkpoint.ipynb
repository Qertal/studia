# (i) year-wise seasonal (like ggseasonplot)
# We'll reshape by month and plot each year as a line across months.
def ggseasonplot_like(series: pd.Series, title: str, polar: bool = False):
    s = series.dropna()
    df = pd.DataFrame({
        "year": s.index.year,
        "month": s.index.month,
        "value": s.values
    })
    months = np.arange(1, 13)

    if polar:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection="polar")
        theta = (months - 1) / 12 * 2 * np.pi
        for y, grp in df.groupby("year"):
            vals = grp.set_index("month").reindex(months)["value"].values
            ax.plot(theta, vals)
        ax.set_title(title)
        # Optional: month tick labels (1..12)
        ax.set_xticks((months - 1) / 12 * 2 * np.pi)
        ax.set_xticklabels(months)
        plt.show()
    else:
        fig, ax = plt.subplots()
        for y, grp in df.groupby("year"):
            vals = grp.set_index("month").reindex(months)["value"].values
            ax.plot(months, vals, alpha=0.8)
        ax.set_title(title)
        ax.set_xlabel("Miesiąc")
        ax.set_ylabel("mln USD")
        ax.set_xticks(months)
        plt.show()

# (ii) Subseries plot (like ggsubseriesplot): one panel per month across years
def ggsubseriesplot_like(series: pd.Series, title: str):
    s = series.dropna()
    df = pd.DataFrame({"year": s.index.year, "month": s.index.month, "value": s.values})
    # 12 subplots (3x4)
    fig, axes = plt.subplots(3, 4, figsize=(12, 7), sharey=True)
    axes = axes.ravel()
    for m in range(1, 13):
        ax = axes[m-1]
        vals = df[df["month"] == m].sort_values("year")
        ax.plot(vals["year"], vals["value"], marker="o")
        ax.set_title(f"Miesiąc {m}")
        ax.set_xlabel("Rok")
        if m in [1,5,9]:
            ax.set_ylabel("mln USD")
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# ---------- 1d) Lag plots (like gglagplot) ----------
def lag_scatter(series: pd.Series, lags=(1,2,3,4), title="Lag-scatter X(t) vs X(t-k)"):
    s = series.dropna()
    L = len(lags)
    cols = min(3, L)
    rows = int(np.ceil(L / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(4*cols+1, 3.6*rows))
    axes = np.atleast_1d(axes).ravel()
    for ax, k in zip(axes, lags):
        x = s.iloc[k:]
        y = s.shift(k).dropna().iloc[k*0:]  # align
        y = s.shift(k).dropna()
        # Align indexes
        xy = pd.concat([x, y], axis=1).dropna()
        ax.scatter(xy.iloc[:, 1], xy.iloc[:, 0], s=12)
        # 45-degree reference
        mn = np.nanmin(xy.values)
        mx = np.nanmax(xy.values)
        ax.plot([mn, mx], [mn, mx], lw=1)
        ax.set_title(f"k = {k}")
        ax.set_xlabel("X(t-k)")
        ax.set_ylabel("X(t)")
    for ax in axes[len(lags):]:
        ax.axis("off")
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# c) Convolution filters (to mirror R: filter(..., method='convolution'))
# R 'sides=2' means centered (non-causal), 'sides=1' means causal.
# We'll use np.convolve in 'same' mode for centered; and a custom causal conv.

def conv_centered(x: pd.Series, h: np.ndarray) -> pd.Series:
    """Centered convolution (non-causal): result ~ sum h[j] * x[t + shift_j].
    We'll align like R's sides=2, method='convolution'. NaNs at edges.
    """
    y = np.convolve(x.values, h[::-1], mode="same")  # reverse h to mimic R's convention
    out = pd.Series(y, index=x.index, dtype=float)
    # Edge effects: mark the first/last floor(len(h)/2) as NaN
    pad = len(h) // 2
    out.iloc[:pad] = np.nan
    out.iloc[-pad:] = np.nan
    return out

def conv_causal(x: pd.Series, h: np.ndarray) -> pd.Series:
    """Causal convolution: X2(t) = sum_{j=0..m-1} h[j] * X(t - j).
    We'll compute via direct loop (efficient enough for lab scale).
    """
    arr = x.values.astype(float)
    m = len(h)
    out = np.full_like(arr, np.nan, dtype=float)
    for t in range(len(arr)):
        acc = 0.0
        valid = True
        for j in range(m):
            idx = t - j
            if idx < 0 or np.isnan(arr[idx]):
                valid = False; break
            acc += h[j] * arr[idx]
        if valid:
            out[t] = acc
    return pd.Series(out, index=x.index, dtype=float)

# I: non-causal example: h(-1)=1/3; h(0)=1/7; h(1)=1/4
# In our centered conv we pass h ordered as [h(-1), h(0), h(1)] and reverse inside.
h_nc = np.array([1/3, 1/7, 1/4], dtype=float)
X1 = conv_centered(X, h_nc)
print("X1 (first 6):")
print(X1.head(6))

# II: causal example: h(0)=1/3; h(1)=1/7; h(2)=1/4  → X2(t)=1/3*X(t)+1/7*X(t-1)+1/4*X(t-2)
h_c = np.array([1/3, 1/7, 1/4], dtype=float)
X2 = conv_causal(X, h_c)
print("\nX2 (first 6):")
print(X2.head(6))

# d) Manual check like in R
chk = pd.concat([X.head(6).rename("X"),
                 X1.head(6).rename("X1"),
                 X2.head(6).rename("X2")], axis=1)

print(chk)

# For centered filter, around t=3 (0-based index 2), expect:
# X1[3] ~ 1/4*X[2] + 1/7*X[1] + 1/3*X[0] (careful with indexing alignment)

```python
# Time Series Lab 2 — Python port (ver. with FRED data)
# Topics: plotting, linear filters, additive decomposition
# Data: TTLCON (Total Construction Spending, USA, monthly)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose
import datetime as dt

# ---------- 0) Settings ----------
plt.rcParams["figure.figsize"] = (9, 4.5)
plt.rcParams["axes.grid"] = True

# ---------- 1) Load TTLCON from FRED ----------
start = dt.datetime(2006, 10, 1)
end = dt.datetime(2025, 10, 1)

X = pdr.DataReader("TTLCON", "fred", start, end)
X = X.asfreq("MS")        # monthly frequency
X = X.rename(columns={"TTLCON": "Wydatki"})
X.index.name = "Time"
print(X.head())

# ---------- 1a) Plot original series ----------
X["Wydatki"].plot(lw=2, color="blue", title="Empiryczny szereg czasowy: wydatki na budownictwo w USA")
plt.xlabel("Czas")
plt.ylabel("X(t) [mln USD]")
plt.show()

# ---------- 2) Seasonal visualizations ----------
# a) Seasonal subseries (month plot)
from statsmodels.graphics.tsaplots import month_plot
month_plot(X["Wydatki"])
plt.title("Wykres sezonowy z rozbiciem na miesiące (Month Plot)")
plt.ylabel("mln USD")
plt.show()

# b) Yearly seasonal plot (custom)
df = X.copy()
df["year"] = df.index.year
df["month"] = df.index.month

months = np.arange(1, 13)
fig, ax = plt.subplots()
for y, grp in df.groupby("year"):
    vals = grp.set_index("month")["Wydatki"].reindex(months)
    ax.plot(months, vals, alpha=0.8, label=y)
ax.set_title("Wykres sezonowy dla poszczególnych lat (d=12)")
ax.set_xlabel("Miesiąc")
ax.set_ylabel("mln USD")
plt.show()

# ---------- 3) Lag plots ----------
def lag_scatter(series: pd.Series, lags=(1,2,3,4), title="Lag scatter"):
    s = series.dropna()
    fig, axes = plt.subplots(1, len(lags), figsize=(4*len(lags), 4))
    if len(lags) == 1:
        axes = [axes]
    for ax, k in zip(axes, lags):
        xy = pd.concat([s, s.shift(k)], axis=1).dropna()
        xy.columns = ["X(t)", f"X(t-{k})"]
        ax.scatter(xy[f"X(t-{k})"], xy["X(t)"], s=12)
        ax.set_title(f"k = {k}")
        ax.set_xlabel(f"X(t-{k})")
        ax.set_ylabel("X(t)")
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

X_cut = X["Wydatki"]["2015-01":"2022-12"]
lag_scatter(X_cut, lags=(1,2,3,4), title="Lag plot: k=1,2,3,4")

# ---------- 4) Differencing ----------
Y = X["Wydatki"].diff()
Y.plot(title="Szereg różnicowany z krokiem 1 (X(t)-X(t-1))")
plt.show()

V = X["Wydatki"].diff(12)
V.plot(title="Szereg po usunięciu sezonowości (X(t)-X(t-12))")
plt.show()

Y_exp = X["Wydatki"].diff().diff(12)
Y_exp.plot(title="Szereg różnicowany eksperymentalnie (1 i 12)")
plt.show()

plot_acf(Y_exp.dropna(), lags=30)
plt.title("ACF(Y_exp)")
plt.show()

# ---------- 5) Linear filters ----------
def conv_causal(x: pd.Series, h: np.ndarray) -> pd.Series:
    arr = x.values.astype(float)
    m = len(h)
    out = np.full_like(arr, np.nan, dtype=float)
    for t in range(len(arr)):
        acc = 0.0
        valid = True
        for j in range(m):
            idx = t - j
            if idx < 0 or np.isnan(arr[idx]):
                valid = False
                break
            acc += h[j] * arr[idx]
        if valid:
            out[t] = acc
    return pd.Series(out, index=x.index, dtype=float)

# Example causal filter: h(0)=1/3; h(1)=1/7; h(2)=1/4
h_c = np.array([1/3, 1/7, 1/4])
X2 = conv_causal(X["Wydatki"], h_c)
print(X2.head())

# ---------- 6) Classical additive decomposition ----------
dec = seasonal_decompose(X["Wydatki"], model="additive", period=12)
trend = dec.trend
seasonal = dec.seasonal
resid = dec.resid

fig, axes = plt.subplots(2, 2, figsize=(12, 7))
axes = axes.ravel()

X["Wydatki"].plot(ax=axes[0], lw=2, title="Dane empiryczne — wydatki na budownictwo w USA")
trend.plot(ax=axes[1], lw=2, title="Oszacowany trend")
seasonal.plot(ax=axes[2], lw=2, title="Komponenta sezonowa")
axes[2].axhline(0, lw=1)
resid.plot(ax=axes[3], lw=2, title="Zaszumienie losowe")
axes[3].axhline(0, lw=1)
plt.tight_layout()
plt.show()
```
