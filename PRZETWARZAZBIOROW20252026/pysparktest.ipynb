```python
from pyspark.sql import SparkSession

import os
os.environ["PYSPARK_PYTHON"] = r"C:\REPO\studia\venv\Scripts\python.exe"
os.environ["PYSPARK_DRIVER_PYTHON"] = r"C:\REPO\studia\venv\Scripts\python.exe"


# Utwórz lokalną sesję Spark
spark = SparkSession.builder \
    .appName("TestSpark") \
    .master("local[*]") \
    .getOrCreate()

# Przykładowe dane
data = [("Ala", 25), ("Ola", 30), ("Jan", 40)]
df = spark.createDataFrame(data, ["Imię", "Wiek"])

# Wyświetl DataFrame
df.show()

# Zatrzymaj sesję
spark.stop()
```

```python
import os
VENV_PY = r"C:\REPO\studia\venv\Scripts\python.exe"  # <-- Twoja ścieżka

os.environ["PYSPARK_PYTHON"] = VENV_PY
os.environ["PYSPARK_DRIVER_PYTHON"] = VENV_PY

from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("LocalTest")
    .master("local[*]")
    .config("spark.driver.bindAddress", "127.0.0.1")
    .config("spark.driver.host", "127.0.0.1")
    .config("spark.python.use.daemon", "false")
    .config("spark.local.dir", r"C:\spark_tmp")   # folder bez spacji, pełne prawa
    .getOrCreate()
)

print("Python exec used by Spark:", spark.sparkContext.pythonExec)
```

```python
import os, sys
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .master("local[*]")
    .config("spark.ui.showConsoleProgress", "false")
    .config("spark.driver.bindAddress", "127.0.0.1")
    .config("spark.driver.host", "127.0.0.1")
    .getOrCreate()
)

data = [("Ala", 23), ("Ola", 19)]
df = spark.createDataFrame(data, ["imie", "wiek"])  # nazwa kolumny dowolna, ale nie ma to wpływu na błąd
df.show()

spark.stop()
```
