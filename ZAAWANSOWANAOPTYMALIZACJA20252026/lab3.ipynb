# ðŸ”· Metoda Nelderaâ€“Meada â€“ intuicyjne wyjaÅ›nienie

Metoda **Nelderaâ€“Meada** to jedna z najprostszych i najbardziej intuicyjnych metod **optymalizacji bez uÅ¼ycia pochodnych**.  
Oznacza to, Å¼e potrafi znaleÅºÄ‡ minimum funkcji **nawet wtedy, gdy nie umiemy obliczyÄ‡ jej gradientu** (pochodnych czÄ…stkowych).  
Dlatego czÄ™sto stosuje siÄ™ jÄ… w praktyce â€“ w fizyce, uczeniu maszynowym czy dopasowywaniu modeli.

---

## ðŸ§© PomysÅ‚ ogÃ³lny

Zamiast poruszaÄ‡ siÄ™ wzdÅ‚uÅ¼ kierunku gradientu (jak w metodach gradientowych), Nelderâ€“Mead uÅ¼ywa **grupy punktÃ³w** zwanej **simpleksem**.

- Dla funkcji dwÃ³ch zmiennych (x, y) simpleks to **trÃ³jkÄ…t**.  
- Dla trzech zmiennych â€“ **czworoÅ›cian**,  
- a ogÃ³lnie: dla $ n $ zmiennych to figura o $ n + 1 $ wierzchoÅ‚kach.

KaÅ¼dy wierzchoÅ‚ek to **punkt w przestrzeni rozwiÄ…zaÅ„**, w ktÃ³rym obliczamy wartoÅ›Ä‡ funkcji celu $ f(x) $.

---

## ðŸ” Jak dziaÅ‚a metoda krok po kroku

1. **Inicjalizacja simpleksu**  
   - Wybieramy punkt startowy $ x_0 $ i tworzymy wokÃ³Å‚ niego pozostaÅ‚e $ n $ wierzchoÅ‚kÃ³w (np. lekko przesuniÄ™tych).  
   - Otrzymujemy $ n+1 $ punktÃ³w: $ x_1, x_2, \dots, x_{n+1} $.  
   - Dla kaÅ¼dego obliczamy wartoÅ›Ä‡ funkcji $ f(x_i) $.

2. **Sortowanie punktÃ³w**  
   - PorzÄ…dkujemy punkty od najlepszego (najmniejsza wartoÅ›Ä‡ $ f $) do najgorszego (najwiÄ™ksza wartoÅ›Ä‡ $ f $).  
   - Oznaczamy:
     - $ x_{\text{best}} $ â€“ najlepszy punkt,
     - $ x_{\text{worst}} $ â€“ najgorszy punkt,
     - $ x_{\text{centroid}} $ â€“ Å›rodek pozostaÅ‚ych punktÃ³w (bez najgorszego).

3. **Odbicie (reflection)**  
   - Chcemy sprawdziÄ‡, czy po drugiej stronie Å›rodka nie ma lepszego miejsca.  
   - Tworzymy nowy punkt:
     $$
     x_r = x_{\text{centroid}} + \alpha (x_{\text{centroid}} - x_{\text{worst}})
     $$
     gdzie $ \alpha > 0 $ to wspÃ³Å‚czynnik odbicia (zwykle 1).  
   - JeÅ›li $ f(x_r) $ jest lepsze niÅ¼ wiÄ™kszoÅ›Ä‡ punktÃ³w, to zastÄ™pujemy $ x_{\text{worst}} $ nowym punktem.

4. **Ekspansja (expansion)**  
   - JeÅ›li odbicie daÅ‚o **duÅ¼o lepszy wynik** (czyli nowy punkt jest nawet lepszy niÅ¼ dotychczasowy najlepszy),  
     to prÃ³bujemy iÅ›Ä‡ jeszcze dalej w tym kierunku:
     $$
     x_e = x_{\text{centroid}} + \gamma (x_r - x_{\text{centroid}})
     $$
     gdzie $ \gamma > 1 $ (np. 2).  
   - JeÅ›li ekspansja przynosi poprawÄ™, zostawiamy $ x_e $; w przeciwnym razie pozostajemy przy $ x_r $.

5. **Kontrakcja (contraction)**  
   - JeÅ›li odbicie **nie pomogÅ‚o** (czyli nowy punkt jest gorszy niÅ¼ dotychczasowe),  
     cofamy siÄ™ w stronÄ™ Å›rodka:
     $$
     x_c = x_{\text{centroid}} + \rho (x_{\text{worst}} - x_{\text{centroid}})
     $$
     gdzie $ 0 < \rho < 1 $ (np. 0.5).  
   - JeÅ›li ten punkt jest lepszy niÅ¼ najgorszy, przyjmujemy go;  
     w przeciwnym razie przechodzimy do nastÄ™pnego kroku.

6. **Redukcja (shrink)**  
   - JeÅ›li nic nie zadziaÅ‚aÅ‚o, zmniejszamy caÅ‚y simpleks wokÃ³Å‚ najlepszego punktu:
     $$
     x_i = x_{\text{best}} + \sigma (x_i - x_{\text{best}}), \quad \text{dla } i = 2, \dots, n+1
     $$
     gdzie $ \sigma \approx 0.5 $.  
   - DziÄ™ki temu wszystkie punkty zbliÅ¼ajÄ… siÄ™ do najlepszego, a obszar poszukiwaÅ„ siÄ™ â€žkurczyâ€.

7. **Kryterium stopu**  
   - Algorytm zatrzymuje siÄ™, gdy:
     - wszystkie punkty simpleksu sÄ… bardzo blisko siebie,  
     - lub rÃ³Å¼nice wartoÅ›ci funkcji $ f(x_i) $ sÄ… bardzo maÅ‚e,  
     - albo osiÄ…gniÄ™to maksymalnÄ… liczbÄ™ iteracji.

---

## âš™ï¸ Co robi metoda w praktyce

Metoda Nelderaâ€“Meada **porusza siÄ™ i deformuje simpleks**:  
czasem go **odbija**, czasem **rozciÄ…ga**, a czasem **kurczy**, szukajÄ…c coraz lepszego obszaru.  
DziÄ™ki temu **nie wymaga Å¼adnych pochodnych** â€“ wystarczy umieÄ‡ obliczyÄ‡ wartoÅ›Ä‡ funkcji $ f(x) $.

---

## ðŸ“‰ Zalety

âœ… DziaÅ‚a nawet dla funkcji:
- niegÅ‚adkich (bez pochodnych),
- zaszumionych lub trudnych do opisania analitycznie,  
- z nieznanymi ograniczeniami.

âœ… Åatwa do zaimplementowania.  
âœ… DziaÅ‚a w dowolnej liczbie wymiarÃ³w.

---

## âš ï¸ Wady

âŒ MoÅ¼e **utknÄ…Ä‡ w minimum lokalnym** (nie zawsze znajduje globalne).  
âŒ Bywa **wolna**, szczegÃ³lnie przy duÅ¼ej liczbie zmiennych.  
âŒ Nie gwarantuje zbieÅ¼noÅ›ci dla wszystkich typÃ³w funkcji.

---

## ðŸ§  Podsumowanie

Metoda Nelderaâ€“Meada to **inteligentne testowanie wartoÅ›ci funkcji** w kilku punktach jednoczeÅ›nie.  
Zamiast â€žzjeÅ¼dÅ¼aÄ‡ w dÃ³Å‚â€ po gradientach, buduje ona i przesuwa **wielokÄ…t (simpleks)** po krajobrazie funkcji,  
aÅ¼ znajdzie miejsce, w ktÃ³rym wartoÅ›ci funkcji przestajÄ… siÄ™ znaczÄ…co poprawiaÄ‡.

W skrÃ³cie:

| Etap | Co siÄ™ dzieje | WspÃ³Å‚czynnik |
|------|----------------|---------------|
| Odbicie | PrzesuniÄ™cie w stronÄ™ lepszego obszaru | Î± â‰ˆ 1 |
| Ekspansja | IdÅº dalej, jeÅ›li lepiej | Î³ â‰ˆ 2 |
| Kontrakcja | Cofnij siÄ™, jeÅ›li gorzej | Ï â‰ˆ 0.5 |
| Redukcja | Zmniejsz caÅ‚y simpleks | Ïƒ â‰ˆ 0.5 |

---

ðŸ“Œ **PodsumowujÄ…c:**  
Metoda Nelderaâ€“Meada to prosty, intuicyjny i bardzo uniwersalny sposÃ³b szukania minimum funkcji.  
Nie potrzebuje pochodnych, wiÄ™c nadaje siÄ™ do problemÃ³w, w ktÃ³rych jedyne, co umiemy zrobiÄ‡, to **zmierzyÄ‡ wartoÅ›Ä‡ funkcji w danym punkcie**.


```python
def rosenbrock(x):
    """f(x1,x2) = (1-x1)^2 + 100*(x2 - x1^2)^2"""
    x = np.asarray(x, dtype=float)
    return (1 - x[0])**2 + 100.0 * (x[1] - x[0]**2)**2

def himmelblau(x):
    """Funkcja Himmelblaua."""
    x, y = x
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2
```

```python
def nelder_mead(f, x0, alpha=1.0, gamma=2.0, rho=0.5, sigma=0.5,
                tol=1e-6, max_iter=500):
    """
    Bardzo prosta implementacja metody Nelderaâ€“Meada (bez pochodnych).
    Zwraca (x_best, f_best, trajektoria, simplex_hist).
    """
    x0 = np.asarray(x0, dtype=float)
    n = x0.size

    simplex = [x0]
    step = 0.05  
    for i in range(n):
        x = x0.copy()
        x[i] = x[i] + step if x[i] != 0 else step
        simplex.append(x)
    simplex = np.array(simplex)

    path = [simplex[0].copy()]
    simplex_hist = [simplex.copy()]  # zapamiÄ™tuj historiÄ™ simpleksu

    for _ in range(max_iter):
        # sortuj wierzchoÅ‚ki po wartoÅ›ci f
        simplex = np.array(sorted(simplex, key=f))
        f_vals = np.array([f(x) for x in simplex])

        # kryterium stopu: rozrzut wartoÅ›ci f
        if f_vals.std() < tol:
            break

        x_best = simplex[0]
        x_worst = simplex[-1]
        centroid = simplex[:-1].mean(axis=0)

        # odbicie
        x_r = centroid + alpha * (centroid - x_worst)
        f_r = f(x_r)

        if f_r < f_vals[0]:
            # ekspansja
            x_e = centroid + gamma * (x_r - centroid)
            simplex[-1] = x_e if f(x_e) < f_r else x_r
        elif f_vals[0] <= f_r < f_vals[-2]:
            # akceptuj odbicie
            simplex[-1] = x_r
        else:
            # kontrakcja
            x_c = centroid + rho * (x_worst - centroid)
            if f(x_c) < f_vals[-1]:
                simplex[-1] = x_c
            else:
                # redukcja do najlepszego
                for i in range(1, n+1):
                    simplex[i] = x_best + sigma * (simplex[i] - x_best)

        path.append(simplex[0].copy())
        simplex_hist.append(simplex.copy())

    x_best = simplex[0]
    return x_best, f(x_best), np.array(path), np.array(simplex_hist)    
```

# Metoda Gradientowa

```python
def rosenbrock(x):
    """Funkcja Rosenbrocka f(x,y) = (1-x)^2 + 100*(y-x^2)^2"""
    x1, x2 = x
    return (1 - x1)**2 + 100 * (x2 - x1**2)**2

def himmelblau(x):
    """Funkcja Himmelblaua."""
    x, y = x
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

def rosenbrock_grad(x):
    """Gradient funkcji Rosenbrocka"""
    x1, x2 = x
    df_dx1 = -2*(1 - x1) - 400*x1*(x2 - x1**2)
    df_dx2 = 200*(x2 - x1**2)
    return np.array([df_dx1, df_dx2])

def himmelblau_grad(x):
    """Gradient funkcji Himmelblaua"""
    x, y = x
    df_dx = 4*x*(x**2 + y - 11) + 2*(x + y**2 - 7)
    df_dy = 2*(x**2 + y - 11) + 4*y*(x + y**2 - 7)
    return np.array([df_dx, df_dy])

def gradient_descent(f, grad_f, x0, lr=0.001, tol=1e-6, max_iter=10000):
    """
    Prosta implementacja metody spadku gradientowego (gradient descent)
    f       - funkcja celu
    grad_f  - gradient funkcji
    x0      - punkt startowy (np.array)
    lr      - learning rate (krok uczenia)
    tol     - tolerancja bÅ‚Ä™du
    max_iter - maks. liczba iteracji
    """
    x = np.array(x0, dtype=float)
    path = [x.copy()]  # historia punktÃ³w

    for i in range(max_iter):
        grad = grad_f(x)
        x_new = x - lr * grad
        path.append(x_new.copy())

        # warunek stopu
        if np.linalg.norm(x_new - x) < tol:
            break
        x = x_new

    return x, f(x), np.array(path)
```

```python
x0 = np.array([-1.5, 2.0])  # start z gÃ³rnej czÄ™Å›ci doliny
x_min, f_min, traj = gradient_descent(rosenbrock, rosenbrock_grad, x0,
                                          lr=0.001, max_iter=20000)
print("Minimum znalezione w:", x_min)
print("WartoÅ›Ä‡ funkcji:", f_min)
plt.plot(traj[:, 0], traj[:, 1], marker='o')
plt.title("Trajektoria spadku gradientowego")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x0 = np.array([0.0, 0.0])  # start blisko jednego z minimÃ³w
x_min, f_min, traj = gradient_descent(himmelblau, himmelblau_grad, x0,
                                          lr=0.001, max_iter=20000)
print("Minimum znalezione w:", x_min)
print("WartoÅ›Ä‡ funkcji:", f_min)
plt.plot(traj[:, 0], traj[:, 1], marker='o')
plt.title("Trajektoria spadku gradientowego")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
